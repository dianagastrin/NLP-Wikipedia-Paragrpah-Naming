```python
import pandas
def splits_list(bio,ind,acc):
    if ind == len(bio)-1:
        return acc
    elif ind == 0:
        acc.append(len(bio[ind][1].split()))
    else:
        acc.append(acc[ind-1]+len(bio[ind][1].split()))
    return splits_list(bio,ind+1,acc)
    
#print(processed1000[1],len(processed1000[1][0][1][0][1].split()))
#print(splits_list(processed1000[1][0][1],0,[]))
data = []
for i,bio in enumerate(processed1000[:100]):
    biotext = bio[0][1]
    onepiece = " ".join([part[1] for part in biotext]).replace('\n','')
    if(len(biotext) > 1):
        data.append({
            'gold' : splits_list(biotext,0,[]),
            'test' : segmentize(onepiece,len(biotext) - 2)
        })
segmentation_with_embeddings = pandas.DataFrame(data)
segmentation_with_embeddings
```
